```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.expected_conditions import *
import time
import os
import json

def open_driver(path,url):
    options = webdriver.ChromeOptions()
    options.add_experimental_option('excludeSwitches', ['enable-logging']) 
    driver = webdriver.Chrome(path,options=options)
    driver.get(url)
    driver.implicitly_wait(7)
    return driver

def search_a_keyword(keyword):
    search_bar = driver.find_element(By.NAME,"q") 
    search_bar.send_keys(keyword) 
    time.sleep(operate_delay)
    search_bar.send_keys(Keys.RETURN) 
    time.sleep(operate_delay)

def get_all_product_price_in_one_page():
    card_bodys = wait.until(presence_of_all_elements_located((By.CLASS_NAME, "card-body")))
    for card_body in card_bodys:
        product_link = card_body.find_element(By.TAG_NAME,'a')
        product_link.send_keys(Keys.CONTROL,Keys.RETURN)
        
        windows = driver.window_handles
        driver.switch_to.window(windows[-1])
        list_items = wait.until(presence_of_all_elements_located((By.CLASS_NAME, "list-group-item")))
        time.sleep(operate_delay)
        
        product = {}
        name = list_items[0].text
        price = (float(''.join(filter(str.isdigit, list_items[2].text))))/100
        product[name]=price
        products.append(product)
        driver.close()
        time.sleep(operate_delay)
        driver.switch_to.window(windows[0])   

def get_price(product):
    for name, price in product.items():
        return price

def save_the_result(result):
    with open('result.json','w') as f:
        json.dump(result,f)

driver_path = './chromedriver.exe' #The path where you put the chromedriver.exe in.
url = 'http://10.113.178.219' #The url of shopping cart. Please connect to UM campus network to run.
page0 = [url]
keyword = input('Please input a keyword to search:')
driver = open_driver(driver_path,url)
wait = WebDriverWait(driver, 10)

elems = driver.find_elements_by_xpath("//ul[@class='pagination']//a[@href]")
page = []
for elem in elems:
    urls=elem.get_attribute('href')
    page.append(urls)
pages = page0+page

if keyword!='':
    for i in pages:
        operate_delay = 0 
        driver = open_driver(driver_path,url) 
        wait = WebDriverWait(driver, 10)
        search_a_keyword(keyword)
        products = []
        get_all_product_price_in_one_page()
        driver.close()
    result = sorted(products, key=get_price,reverse=True)[:5]
    print(result)
    save_the_result(result)
else:
    all_product = []
    for i in range(1,10):
        url = 'http://10.113.178.219/page/%d'%i
        operate_delay = 0 
        driver = open_driver(driver_path,url) 
        wait = WebDriverWait(driver, 10)
        products = []
        get_all_product_price_in_one_page()
        driver.close()
        all_product.append(products)    
    lists = sum(all_product,[])
    result = sorted(lists,key=get_price,reverse=True)[:5]
    print(result)
    save_the_result(result)
```
